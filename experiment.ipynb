```python
df.drop(['CustomerID','InvoiceDate','StockCode','InvoiceNo'],axis=1,inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.dropna(subset=['Description'], inplace=True)
```

```python
df['Quantity'].nunique()
df['Quantity'].unique()
```

```python
df.drop(df[df['Quantity'] < 0].index, inplace=True)
df['Quantity'].unique()
```

```python
df.reset_index(drop=True,inplace=True)
```

```python
df
```

```python
Q1 = df['Quantity'].quantile(0.25)
Q3 = df['Quantity'].quantile(0.75)
IQR = Q3 - Q1 
df = df[(df['Quantity'] >= (Q1 - 1.5 * IQR)) & (df['Quantity'] <= (Q3 + 1.5 * IQR))]
```

```python
Q1 = df['UnitPrice'].quantile(0.25)
Q3 = df['UnitPrice'].quantile(0.75)
IQR = Q3 - Q1
df = df[(df['UnitPrice'] >= (Q1 - 1.5 * IQR)) & (df['UnitPrice'] <= (Q3 + 1.5 * IQR))]
```

```python
df['Total Price']=df['Quantity']*df['UnitPrice']
```

```python
df.to_csv('cleaned_data.csv', index=False)
```

```python
cleaned_df=pd.read_csv('cleaned_data.csv')
```

```python
cleaned_df.head(7)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
# Select the features for clustering
features = ['Quantity', 'UnitPrice', 'Total Price']
X = cleaned_df[features]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
# Set the number of clusters
kmeans = KMeans(n_clusters=3, random_state=42)
cleaned_df['Cluster'] = kmeans.fit_predict(X_scaled)
print("Data with clusters assigned:")
cleaned_df[['Description', 'Quantity', 'UnitPrice', 'Total Price', 'Cluster']]
```

```python
# Apply PCA to reduce data to 2D
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
cleaned_df['PCA1'] = X_pca[:, 0]
cleaned_df['PCA2'] = X_pca[:, 1]
plt.figure(figsize=(8, 6))
plt.scatter(cleaned_df['PCA1'], cleaned_df['PCA2'], c=cleaned_df['Cluster'])
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.title('K-Means Clustering with PCA')
plt.colorbar(label='Cluster')
plt.show()
```

```python
#optimal value for k using elbow method
wcss = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)  
    wcss.append(kmeans.inertia_) 
```

```python
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('WCSS (Within-Cluster Sum of Squares)')
plt.grid(True)
plt.show()
```
